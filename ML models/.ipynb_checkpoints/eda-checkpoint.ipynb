## 1.6 Check the outliers

```python
import pandas as pd

# Create an empty dictionary to store outliers
outliers_dict = {}

# Loop through all numeric columns
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    # Store the outliers
    outliers_dict[col] = outliers

    print(f"\n🔍 Outliers in '{col}':")
    print(outliers[[col]].head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create an empty dictionary to store outliers
outliers_dict = {}

# Set plotting style
sns.set(style="whitegrid")

# Loop through all numeric columns
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers_dict[col] = outliers

    print(f"\n🔍 Outliers in '{col}':")
    print(outliers[[col]])

    # --- Visualization for each column ---
    fig, axs = plt.subplots(1, 3, figsize=(18, 4))
    fig.suptitle(f'Outlier Detection for "{col}"', fontsize=16)

    # 1. Boxplot
    sns.boxplot(x=df[col], ax=axs[0], color='skyblue')
    axs[0].set_title('Box Plot')

    # 2. Scatter plot with outliers highlighted
    axs[1].scatter(df.index, df[col], label='Data', alpha=0.6)
    axs[1].scatter(outliers.index, outliers[col], color='red', label='Outliers')
    axs[1].set_title('Scatter Plot')
    axs[1].legend()

    # 3. Distribution plot
    sns.histplot(df[col], kde=True, ax=axs[2], color='green')
    axs[2].axvline(lower_bound, color='red', linestyle='--')
    axs[2].axvline(upper_bound, color='red', linestyle='--')
    axs[2].set_title('Distribution with IQR Bounds')

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd

# Create a copy of the DataFrame to preserve the original
df_no_outliers = df.copy()

# Loop through all numeric columns and filter out outliers
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Keep only values within the IQR bounds
    df_no_outliers = df_no_outliers[
        (df_no_outliers[col] >= lower_bound) & (df_no_outliers[col] <= upper_bound)
    ]

# Show the shape and a preview of the cleaned data
print("✅ Data shape after removing outliers:", df_no_outliers.shape)
df_no_outliers.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your original DataFrame

# Create an empty dictionary to store outliers
outliers_dict = {}

# Set plotting style
sns.set(style="whitegrid")

# Loop through all numeric columns for outlier detection and visualization
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers for this column
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers_dict[col] = outliers

    print(f"\n🔍 Outliers in '{col}':")
    print(outliers[[col]])

    # Visualization
    fig, axs = plt.subplots(1, 3, figsize=(18, 4))
    fig.suptitle(f'Outlier Detection for "{col}"', fontsize=16)

    # Boxplot
    sns.boxplot(x=df[col], ax=axs[0], color='skyblue')
    axs[0].set_title('Box Plot')

    # Scatter plot with outliers highlighted
    axs[1].scatter(df.index, df[col], label='Data', alpha=0.6)
    axs[1].scatter(outliers.index, outliers[col], color='red', label='Outliers')
    axs[1].set_title('Scatter Plot')
    axs[1].legend()

    # Distribution plot with IQR bounds
    sns.histplot(df[col], kde=True, ax=axs[2], color='green')
    axs[2].axvline(lower_bound, color='red', linestyle='--')
    axs[2].axvline(upper_bound, color='red', linestyle='--')
    axs[2].set_title('Distribution with IQR Bounds')

    plt.tight_layout()
    plt.show()

# Create a copy of the DataFrame to preserve original data
df_no_outliers = df.copy()

# Loop through numeric columns to remove outliers (keep only data within bounds for all columns)
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_no_outliers = df_no_outliers[
        (df_no_outliers[col] >= lower_bound) & (df_no_outliers[col] <= upper_bound)
    ]

# Show the shape and preview of the cleaned DataFrame
print("✅ Data shape after removing outliers:", df_no_outliers.shape)
display(df_no_outliers.head())

```
