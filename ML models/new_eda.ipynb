# 1. Bar Plot EDA

## In Univariate analysis plotting the right charts can help us to better understand the data making the data visualization so important. 

```python
import matplotlib.pyplot as plt
```

```python
# Count the occurrences of each category in the 'gender' column
gender_counts = df['gender'].value_counts()

# Plotting
plt.figure(figsize=(8,6))
plt.bar(gender_counts.index, gender_counts, color='skyblue')
plt.title('Count Plot of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python


# Count the occurrences of each category in 'part_time_job' column
part_time_job_counts = df['part_time_job'].value_counts()

# Plotting
plt.figure(figsize=(8,6))
plt.bar(part_time_job_counts.index, part_time_job_counts, color='deeppink')
plt.title('Count Plot of Part Time Job')
plt.xlabel('Part Time Job (Yes/No)')
plt.ylabel('Count')
plt.show()
```

```python
# Count the occurrences of each category in the 'diet_quality' column
diet_quality_counts = df['diet_quality'].value_counts()

# Plotting
plt.figure(figsize=(8,6))
plt.bar(diet_quality_counts.index, diet_quality_counts, color='mediumseagreen')
plt.title('Count Plot of Diet Quality')
plt.xlabel('Diet Quality')
plt.ylabel('Count')
plt.show()
```

```python
# Count the occurrences of each category in the 'parental_education_level' column
education_counts = df['parental_education_level'].value_counts()

# Plotting
plt.figure(figsize=(10,6))  # Wider figure if many categories
plt.bar(education_counts.index, education_counts, color='cornflowerblue')
plt.title('Count Plot of Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate labels if they are long
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each category in the 'internet_quality' column
internet_quality_counts = df['internet_quality'].value_counts()

# Plotting
plt.figure(figsize=(8,6))
plt.bar(internet_quality_counts.index, internet_quality_counts, color='orange')
plt.title('Count Plot of Internet Quality')
plt.xlabel('Internet Quality')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each category in the 'extracurricular_participation' column
extra_counts = df['extracurricular_participation'].value_counts()

# Plotting
plt.figure(figsize=(8,6))
plt.bar(extra_counts.index, extra_counts, color='purple')
plt.title('Count Plot of Extracurricular Participation')
plt.xlabel('Extracurricular Participation')
plt.ylabel('Count')
plt.show()
```

# 2.  Kernel density plot for understanding variance in the dataset

```python
import seaborn as sns

sns.set_style("darkgrid")

numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

### Conclusion for Kernel Density Plots (KDE) on Your Dataset’s Numerical Features
The KDE plots provide a smooth visualization of the distributions and variance of key numerical variables related to student lifestyle and academic performance:

Study Hours Per Day and Exam Score: Show moderate to high variance, indicating a wide range of study habits and academic outcomes among students.

Social Media Hours and Netflix Hours: Display right-skewed distributions, meaning that while most students spend limited time on these activities, a smaller group spends significantly more time.

Attendance Percentage: Is concentrated near higher values, suggesting that most students maintain good attendance, but there is some spread.

Sleep Hours: Exhibits a fairly normal distribution, indicating relatively consistent sleeping patterns across the students.

Mental Health Rating and Exercise Frequency: Show some variability, reflecting diverse mental wellness and physical activity levels.

Age: Likely clustered around typical student ages, with low variance.

Overall, these KDE plots highlight important differences in how students allocate time and manage various lifestyle factors, which can impact academic performance and well-being. Understanding these distributions helps in identifying potential outliers, subgroups, and areas for targeted intervention or further analysis.
