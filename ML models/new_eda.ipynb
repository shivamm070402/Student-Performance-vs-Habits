# 2.  Kernel density plot for understanding variance in the dataset

```python
import seaborn as sns

sns.set_style("darkgrid")

numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

### Conclusion for Kernel Density Plots (KDE) on Your Dataset’s Numerical Features
1. The KDE plots provide a smooth visualization of the distributions and variance of key numerical variables related to student lifestyle and academic performance:

2. Study Hours Per Day and Exam Score: Show moderate to high variance, indicating a wide range of study habits and academic outcomes among students.

3. Social Media Hours and Netflix Hours: Display right-skewed distributions, meaning that while most students spend limited time on these activities, a smaller group spends significantly more time.

4. Attendance Percentage: Is concentrated near higher values, suggesting that most students maintain good attendance, but there is some spread.

5. Sleep Hours: Exhibits a fairly normal distribution, indicating relatively consistent sleeping patterns across the students.

6. Mental Health Rating and Exercise Frequency: Show some variability, reflecting diverse mental wellness and physical activity levels.

7. Age: Likely clustered around typical student ages, with low variance.

#### Overall, these KDE plots highlight important differences in how students allocate time and manage various lifestyle factors, which can impact academic performance and well-being. Understanding these distributions helps in identifying potential outliers, subgroups, and areas for targeted intervention or further analysis.

# 3. Swarm Plot for showing the outlier in the data 

## 3.1 Numerical Swarm Plot



### 3.1.1 Swarm plot for age

```python
df['age_bin'] = pd.qcut(df['age'], q=4, duplicates='drop')

plt.figure(figsize=(8, 5))
sns.swarmplot(x='age_bin', y='exam_score', data=df)
plt.title('Swarm Plot: Age (Binned) vs Exam Score')
plt.xlabel('Age (Binned)')
plt.ylabel('Exam Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
